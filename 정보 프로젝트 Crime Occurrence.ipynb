```python
import csv
f = open('2016년.csv')

name=input('알고 싶은 지역구를 입력하세요:')
data = csv.reader(f)
next(data)
data = list(data)
crime = []

for row in data :
    if row[0] == name and row[2] == '발생' :
        crime.append(int(row[3]))

import matplotlib.pyplot as plt
plt.figure(figsize=(6,4), dpi=500)
plt.rc('font',family='Malgun Gothic')

size = [crime[0], crime[1], crime[2], crime[3], crime[4]]

labels = '살인','강도','강간','절도','폭력'
colors = ['yellowgreen','gold','lightskyblue','lightcoral','pink']

plt.pie(size, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

plt.axis('equal')
plt.show()
```

```python
import csv
f = open('2016년.csv')

name = input('알고 싶은 지역구를 입력하세요:')
data = csv.reader(f)
next(data)
data = list(data)
crime = []
find = []

for row in data :
    if row[0] == name and row[2] == '발생' :
        crime.append(int(row[3]))
    elif row[0] == name and row[2] == '검거' :
        find.append(int(row[3]))

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4), dpi=500)
plt.rc('font',family='Malgun Gothic')

crimegraph = ['살인','강도','강간','절도','폭력']
occurence = [crime[0],crime[1],crime[2],crime[3],crime[4]]
catch = [find[0],find[1],find[2],find[3],find[4]]

plt.bar(range(0,9,2),occurence, width = 0.4, label = '발생횟수')
plt.bar(range(1,10,2),catch, width = 0.4, label = '검거횟수')
labels = '살인','강도','강간','절도','폭력'

plt.title('요청하진 지역구의 범죄 현황입니다.')

plt.legend()
plt.xticks([0.5,2.5,4.5,6.5,8.5],['살인','강도','강간','절도','폭력'])
```
